Escribir la función cantidadPrimos(arreglo), que retorna la cantidad de números primos que forman parte de un arreglo. Las soluciones que NO sean recursivas no serán tomadas en cuenta.

Por ejemplo:
Si arr1 = [5, 10, 9, 8, 13, 21] => cantidadPrimos(arr1) = 2
Ayuda: Se puede asumir que ya se tiene la función esPrimo(n) que devuelve verdadero cuando n es primo.


```python
import numpy as np

def esPrimo(n):
    divisibles = 0
    for numero in range(1, n+1):
        if n % numero == 0:
            divisibles += 1
    return 1 if divisibles == 2 else 0

def cantidadPrimos(arreglo, pos=0):
    cantidad = 0
    if pos == len(arreglo):
      cantidad = 0
    elif esPrimo(arreglo[pos]):
      cantidad = 1 + cantidadPrimos(arreglo, pos+1)
    else:
      cantidad = cantidadPrimos(arreglo, pos+1)
    return cantidad
```

Cada edificio tiene un nombre y aulas distribuidas en 3 pisos. Vamos a suponer que un edificio de aulas tiene la misma cantidad de aulas en cada uno de los pisos. Lo que se necesita hacer es asignar materias con sus correspondientes estudiantes a las aulas de los edificios.  

Implementar el **TDA EdificioUnaHur** que contenga al menos las siguientes operaciones:

```python
import numpy as np

class Aula:
  def __init__(self, materia, cantAlumnos):
    self.__cantAlumnos = cantAlumnos
    self.__materia = materia

  def __repr__(self):
    return f"materia: {self.__materia}, alumnos: {self.__cantAlumnos} -"

  def getCantAlumnos(self):
    return self.__cantAlumnos

class EdificioUnaHur:
  def __init__(self, nombre, cantAulas):
    self.__nombre = nombre
    self.__estructura = np.empty((cantAulas, 3), Aula)

  def __repr__(self):
    return f"Edificio {self.__nombre} - {self.__estructura}"

  def asignarAula(self, nombreMateria, cantEstudiantes):
    for piso in range(len(self.__estructura)):
      for aula in range(len(self.__estructura[0])):
        if self.__estructura[piso][aula] == None:
          self.__estructura[piso][aula] = Aula(nombreMateria, cantEstudiantes)
          return (piso, aula)
    raise Exception("No hay aulas disponibles")

  def cantidadAlumnos(self, piso):
    alumnos = 0
    for aula in self.__estructura[piso]:
      if aula != None:
          alumnos += aula.getCantAlumnos()
    return alumnos

  def nuevaCursada(self):
    for piso in range(self.__estructura.shape[0]):
      for aula in range(self.__estructura.shape[1]):
        self.__estructura[piso][aula] = None

edificio = EdificioUnaHur("malvinas", 4)
print(edificio)
edificio.asignarAula("EstructuraDatos com1", 45) ; edificio.asignarAula("EstructuraDatos com2", 70) ; edificio.asignarAula("POO 1 com1", 33)
print(edificio)
print(edificio.cantidadAlumnos(0))
edificio.nuevaCursada()
print(edificio)
```
