Por ejemplo:

```python
def inversionRecursiva(string):
  if len(string) == 0:
    return string
  else:
    return inversionRecursiva(string[1:]) + string[0]
```

Crear el TDA **“SalaDeCine”** que modela una sala de cine con los siguientes componentes: una estructura que permita almacenar las butacas, un elemento por cada butaca (por defecto será una sala de 30 filas, con 40 asientos cada una), el número de sala, el tipo (2 posibles: “2D” o “3D”). La sala se va ocupando con espectadores, un espectador se define por dos cosas, dni y edad.

Escribir el constructor del TDA y las siguientes operaciones:

*  **ocupaAsiento(fila,columna, dni, edad)**: Recibe por parámetros la ubicación del espectador que ingresa (fila y columna) y ocupa la
butaca con el espetador.
*  **AsientosVaciosFila(fila)**: Operación que recibe un número de fila y devuelve la cantidad de butacas vacías en esa fila.
*  **gananciaTotal()**: Operación que calcula la ganancia total de la función, teniendo en cuenta que la entrada de los menores de 14 años sale \$2500 y la de los mayores \$4000.
*  **vaciarSala()**: Operación que vacía la sala luego de que termine la película, marcando como desocupada a todas las butacas para la próxima función.

```python
import numpy as np
from enum import Enum
class TipoPelicula(Enum):
  pelicula2D = 1
  pelicula3D = 2

class Espectador:
  def __init__(self, dni, edad):
    self.dni = dni
    self.edad = edad

  def __repr__(self):
    return f"espectador {self.dni}, {self.edad} años -"

class SalaDeCine:
  def __init__(self, numeroSala, tipo):
    self.sala = np.empty((3, 4), object)
    self.numeroSala = numeroSala
    self.tipo = TipoPelicula(tipo)

  def __repr__(self):
    return f"Sala numero {self.numeroSala}, {self.tipo}. Espectadores: {self.sala}"

  def ocupaAsiento(self, fila, butaca, dni, edad):
    if self.sala[fila][butaca] == None:
      self.sala[fila][butaca] = Espectador(dni, edad)
    else:
      return f"la fila {self.fila}, butaca {self.butaca} ya esta ocupada"

  def asientosVaciosFila(self, fila):
    suma = 0
    for butaca in self.sala[fila]:
      suma += 1 if butaca == None else 0
    return suma

  def gananciaTotal(self):
    ganancia = 0
    for fila in self.sala:
      for espectador in fila:
        if espectador != None:
          ganancia += 2500 if espectador.edad <= 14 else 4000
    return ganancia

  def vaciarSala(self):
    for nFila in range(self.sala.shape[0]):
      for nButaca in range(self.sala.shape[1]):
        self.sala[nFila][nButaca] = 0

sala1 = SalaDeCine(1, 1)
e1 = Espectador(4546, 19)
e2 = Espectador(6677, 14)
sala1.ocupaAsiento(0, 1, 4546, 19)
sala1.ocupaAsiento(0, 2, 6677, 14)
print(sala1)
print(sala1.gananciaTotal())
print(sala1.asientosVaciosFila(1))
sala1.vaciarSala()
print(sala1)
```

Escribir la función ***recursiva*** **"sonIguales"** que recibe dos strings del mismo tamaño (no vacías) y retorna *True* si son iguales y *False* en caso contrario.

```python
def validarTamanio(str1, str2):
  if len(str1) != len(str2):
    raise Exception("Los strings deben ser del mismo tamaño")

def sonIguales(str1:str, str2:str, pos=0)-> bool:
  validarTamanio(str1, str2)
  salida = None
  if pos == len(str1):
    salida = True
  elif str1[pos] != str2[pos]:
    salida = False

  else:
    salida = sonIguales(str1, str2, pos+1)

  return salida

```

```python
import copy as cp

class Pila:
  def __init__(self, listaInicial=None):
    self.pila = []
    if listaInicial:
      self.pila = cp.deepcopy(listaInicial)

  def tamanio(self)->int:
    return len(self.pila)

  def estaVacia(self)->bool:
    return self.tamanio() == 0

  def apilar(self, dato:int)->None:
    self.pila.append(dato)

  def tope(self)->int:
    datoCima = None
    if not self.estaVacia():
      datoCima = self.pila[len(self.pila)-1]
    return datoCima

  def desapilar(self)->int:
    dato = None
    if not self.estaVacia():
      dato = self.pila.pop()
    return dato

  def vaciar(self)->None:
    self.pila = []

  def clonar(self):
    return cp.deepcopy(self)

  def posicion(self, pos):
    return self.pila.index(pos)

  def __repr__(self)->str:
    return str(self.pila)+"<->"
```

```python
import numpy as np

def incluidoAux(elemento, arr):
  for i in range(len(arr)):
    if arr[i] == elemento:
      return True
  return False

def incluidoRecursivo(arr1, arr2):
    incluido = None
    if len(arr1) == 0:
      incluido = True
    elif not incluidoAux(arr1[0], arr2):
      incluido = False
    else:
      incluido = incluidoRecursivo(arr1[1:], arr2)
    return incluido

a = np.array([1, 3, 6])
b = np.array([1, 2, 3, 4, 5])
print(incluidoRecursivo(a, b))


def recursivaincluido(primerArreglo, indice, segundoArreglo):
    resultado = False
    if indice == 0:
        resultado = incluidoAux(primerArreglo[indice], segundoArreglo)
    else:
        resultado = recursivaincluido(primerArreglo, indice-1, segundoArreglo) and incluidoAux(primerArreglo[indice], segundoArreglo)
    return resultado  # Añadir esta línea para devolver el resultado

vector = np.array([2, 4, 7])
vector2 = np.array([2, 4, 1, 2, 7, 9])

print(recursivaincluido(vector, len(vector)-1, vector2))
```

```python
import numpy as np
def maximo_arreglo(arr):
    maximo = arr[0]
    if len(arr) == 1:
        maximo = arr[0]
    else:
        maximo = max(maximo, maximo_arreglo(arr[1:]))
    return maximo

# Ejemplo de uso:
arreglo = np.array([3, 8, 1, 6, 2, 541, 9])
print(maximo_arreglo(arreglo))  # Output: 9
```

```python
import numpy as np

def esPar(numero):
  if numero % 2 == 0:
    return True
  else:
    return False

def cantidadPares(arreglo, pos = 0):
  cantidad =0
  if pos == len(arreglo):
    cantidad = 0
  elif esPar(arreglo[pos]):
    cantidad = 1 + cantidadPares(arreglo, pos+1)
  else:
    cantidad = cantidadPares(arreglo, pos+1)
  return cantidad

def cantidadMayores(arreglo, pos = 0):
  cantidad = 0
  if pos == len(arreglo):
    cantidad = 0
  elif arreglo[pos] > 10:
    cantidad = 1 + cantidadMayores(arreglo, pos+1)
  else:
    cantidad = cantidadMayores(arreglo, pos+1)
  return cantidad

def masParesQue10(arreglo):
  return cantidadPares(arreglo) > cantidadMayores(arreglo)

v = np.array([1, 1, 11, 1])

print(cantidadMayores(v))
print(cantidadPares(v))
print(masParesQue10(v))
```

```python

class Auto:
  def __init__(self, patente, horaIngreso):
    self.__patente = patente
    self.__horaIngreso = horaIngreso

  def __repr__(self):
    return f"patente: {self.__patente} - hora ingreso: {self.__horaIngreso}. "

  def getPatente(self):
    return self.__patente

  def getHoraIngreso(self):
    return self.__horaIngreso

class EstacionamientoUnaHur:
  def __init__(self, cantLugares):
    self.__estacionamiento = np.empty((cantLugares, cantLugares), Auto)

  def __repr__(self):
    return f" {self.__estacionamiento}"

  def estacionar(self, patente, horaIngreso):
    for fila in range(len(self.__estacionamiento[0])):
      for lugar in range(len(self.__estacionamiento[1])):
        if self.__estacionamiento[fila][lugar] == None:
          self.__estacionamiento[fila][lugar] = Auto(patente, horaIngreso)
          return fila, lugar
    raise Exception("No hay lugar disponible")

  def salir(self, patente):
    for fila in range(len(self.__estacionamiento[0])):
      for col in range(len(self.__estacionamiento[1])):
        lugar = self.__estacionamiento[fila][col]
        if lugar != None:
          if lugar.getPatente() == patente:
            hora = lugar.getHoraIngreso()
            self.__estacionamiento[fila][col] = None
            return hora

  def estaVacio(self):
    for lugar in self.__estacionamiento:
      for auto in lugar:
        if auto != None:
          return False
    return True

  def cantidadAutosHora(self, hora):
    cantidad = 0
    for lugar in self.__estacionamiento:
      for auto in lugar:
        if auto != None:
          if auto.getHoraIngreso() < hora:
            cantidad += 1
    return cantidad

estacionamiento = EstacionamientoUnaHur(3)
print(estacionamiento)
print(estacionamiento.estacionar(3333, 4))
print(estacionamiento.estacionar(3334, 4))
print(estacionamiento.estacionar(3335, 5))
print(estacionamiento)
print(estacionamiento.salir(3334))
print(estacionamiento)
print(estacionamiento.estacionar(3338, 5))
print(estacionamiento)
print(estacionamiento.cantidadAutosHora(6))
print(estacionamiento.estaVacio())
```
