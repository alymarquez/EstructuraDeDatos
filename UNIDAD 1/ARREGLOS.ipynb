### **Ejercicio 1**

Escribir una función que recibe un numero entero N y le pide al usuario que ingrese N numeros. Al final, el programa debe retornar un vector contiendo los números que fueron ingresados. Tambien debe imprimir por pantalla la suma total de los valores y el promedio. Se deben hacer funciones para resolver el promedio y la suma total.

```
def numerosEnteros(cantidad:int):
  numeros = np.zeros(cantidad, int) # creo un vector de ceros, con el tamaño segun se tome como argumento en el parámetro cantidad

  for indice in range(len(numeros)): # itero sobre el indice del vector (vector[0], vector[1])
    numeros[indice] = int(input("ingrese un numero: ")) # cada numero ingresado se "almacena" en el indice del vector

  print(sumaTotal(numeros))
  print(promedio(numeros))
  return numeros

def sumaTotal(vector:np.array)->int:
  suma = 0
  for elemento in vector:
    suma += elemento
  return suma

def promedio(vector:np.array)->float:
  return sumaTotal(vector) / len(vector)
```

```
def invertirString(vector:np.array)->str:
  string = np.zeros(len(vector), str)
  pos = 0

  for elemento in reversed(vector):
    string[pos] = elemento
    pos += 1
  return string
```

Escribir la función sumaDeVectores que recibe dos vectores de enteros del mismo tamaño y retorna otro con la suma de ambos. La suma de vectores se realiza componente a componente, segun la siguiente forma:

v<sub>1</sub> = [ v<sub>11</sub> , v<sub>12</sub> , v<sub>13</sub> , ... , v<sub>1N</sub> ]

v<sub>2</sub> = [ v<sub>21</sub> , v<sub>22</sub> , v<sub>23</sub> , ... , v<sub>2N</sub> ]

v<sub>1</sub> + v<sub>2</sub> = [ v<sub>11</sub> + v<sub>21</sub> , v<sub>12</sub> + v<sub>22</sub> , v<sub>13</sub> + v<sub>23</sub> , ... , v<sub>1N</sub> + v<sub>2N</sub> ]

```
def sumaDeVectores(v1:np.array, v2:np.array)->int:
  sumaVectores = np.zeros(len(v1), int)

  if len(v1) == len(v2):
    for indice in range(len(sumaVectores)):
      sumaVectores[indice] = v1[indice] + v2[indice]
    return sumaVectores
```

### **Ejercicio 5**

Desarrollar una función que inserte un elemento en una posición en un vector. Al insertar el elemento, se debe producir un desplazamiento a la derecha de todos los elementos en las posiciones siguientes, el último elemento se pierde.

Por ejemplo:

v1 = [2, 4, 1, 7, 6, 2]

**insertar(v1, 5, 3)** #insertamos un 5 en la posición 3

=> v1 = [2, 4, 1, **5**, 7, 6]

```
#Modificar la implementacion del ejercicio 4 pero modificando las posiciones del range

def insertar(vector:np.array, numero:int, posicion:int):

  for indice in reversed(range(posicion, len(vector))):
    vector[indice] = vector[indice - 1]
  vector[posicion] = numero
```

```
v1 = [2, 4, 1, 7, 6, 2]
print(v1)
insertar(v1, 5, 3)
print(v1)
```

Escribir una función que elimine el elemento que ocupa una determinada posición de un vector. Al eliminar, los elementos a la derecha del eliminado, deben desplazarse a la izquierda un lugar y agregar un cero en la última posición.

Por ejemplo:

v1 = [4, 3, **5**, 8, 6, 7]

**eliminar(v1, 2)** #eliminamos el elemento de la posicion 2

=> v1 = [4, 3, **8, 6, 7, 0**]

```
def eliminar(vector:np.array, posicion:int):

  for indice in range(posicion, len(vector)-1):
    vector[indice] = vector[indice + 1]
  vector[-1] = 0
```

### **Ejercicio 11**

Escribir una función **recursiva** que retorna *True* si una cadena de caracteres es un palíndromo y *False* en caso contrario.

Nota: Palíndromo es la forma elegante de "capicua"

```

#salas
def palindromo_recursivo(texto):
    # Caso base: Si la longitud de la cadena es 0 o 1, es un palíndromo
    if len(texto) <= 1:
        return True
    # Caso recursivo: Verificar si el primer y último carácter son iguales
    elif texto[0] == texto[-1]:
        # Llamada recursiva con la cadena sin el primer y último carácter
        return palindromo_recursivo(texto[1:-1])
    else:
        return False
```

```
p = "reconocer"
palindromo_recursivo(p)
```

```
import numpy as np
def palindromo(vector):
    palabraReversed = np.zeros(len(vector), str)
    pos = 0
    for indice in reversed(range(len(palabraReversed))):
        palabraReversed[indice] = vector[pos]
        pos += 1
    if list(palabraReversed) == list(vector):
        esPalindromo = vector + " es palindromo"
    else:
        esPalindromo = vector + " no es palindromo"

    return esPalindromo
```

### **Ejercicio 12**

Escribir una función que recibe las dimensiones y retorna una matriz rellena de la siguiente forma: En la posición ( n , m ) debe contener n + m.

Por ejemplo:

mat1 = rellenaMatriz(4,3)

=> mat1 =

| <!--   --> |Col 0|Col 1|Col 2|
|------------|:---:|:---:|:---:|
| **Fila 0** |  0  |  1  |  2  |
| **Fila 1** |  1  |  2  |  3  |
| **Fila 2** |  2  |  3  |  4  |
| **Fila 3** |  3  |  4  |  5  |

```
def rellenaMatriz(nFilas:int, nCols:int)->np.array:
  matrizSalida = np.zeros((nFilas,nCols), int)
  for nFila in range(nFilas):
    for nCol in range(nCols):
      matrizSalida[nFila, nCol] = nFila+nCol
  return matrizSalida

print(rellenaMatriz(5,5))
```

```
def esCuadrada(nFilas, nCols):
  if nFilas != nCols:
    raise Exception("La matriz no es cuadrada")

def sumaMatrizSinDiagonal(matriz):
  suma = 0
  esCuadrada(matriz.shape[0], matriz.shape[1])

  for nFila in range(matriz.shape[0]):
    for nCol in range(matriz.shape[1]):
      if nCol != nFila:
        suma += matriz[nFila, nCol]
  return suma

m = np.array([[2, 1, 5],
              [1, 2, 3],
              [2, 2, 3]])
sumaMatrizSinDiagonal(m)
```

Es decir:

**Si i, j son los índices de las filas y columnas de M**

**=> Una matriz M es simétrica si:**

**M [ i , j ] == M[ j, i ] para todo i != j**

Matriz simétrica de 4 x 4:

| <!--   --> |Col 0|Col 1|Col 2|Col 3|
|------------|:---:|:---:|:---:|:---:|
| **Fila 0** |  2  |  1  |  0  |  5  |
| **Fila 1** |  1  |  2  |  9  |  7  |
| **Fila 2** |  0  |  9  |  0  |  8  |
| **Fila 3** |  5  |  7  |  8  |  4  |

```
import numpy as np
def matrizSimetricaSinDiagonal(matriz):
  cFilas, cCols = matriz.shape
  esCuadrada(cFilas, cCols)
  esSimetrica = True

  for cFila in range(cFilas):
    for cCol in range(cCols):
      if cFila != cCol:
        esSimetrica = esSimetrica and matriz[cFila, cCol] == matriz[cCol, cFila]

  return esSimetrica


m = np.array([[2, 2, 2],
              [2, 5, 2],
              [2, 2, 7]])
print(matrizSimetricaSinDiagonal(m))
```

```
def sumaDiagonalDeMatriz(matriz, indice = 0):
  suma = 0
  if indice == len(matriz)-1:
    suma = matriz[-1][-1]
  else:
    suma = matriz[indice][indice] + sumaDiagonalDeMatriz(matriz, indice+1)
  return suma

m = np.array([[2, 2, 2],
              [2, 3, 2],
              [2, 2, 3]])
sumaDiagonalDeMatriz(m)
```

```
# Suma de matrices:
# Escribe una función en Python que reciba dos matrices como entrada y devuelva la suma de esas matrices.
# Asegúrate de manejar el caso en el que las matrices tengan diferentes dimensiones.

def verificarDimensiones(m1, m2):
    if len(m1) != len(m2) or len(m1[0]) != len(m2[0]):
      raise Exception("las matrices no tienen las mismas dimensiones")

def sumaDeMatrices(m1, m2):
  suma = np.zeros((len(m1),len(m1)), int)
  nFilas, nCols = suma.shape

  verificarDimensiones(m1, m2)
  for nFila in range(nFilas):
    for nCol in range(nCols):
      suma[nFila, nCol] = m1[nFila, nCol] + m2[nFila, nCol]

  return suma
```

```
m1 = np.array([[3, 3, 3],
                [3, 4, 5],
                [2, 4, 5]])

m2 = np.array([[2, 4, 2],
                [4, 5, 6],
                [2, 7, 8]])

print(sumaDeMatrices(m1, m2))
```

Implementar una función recursiva para encontrar el máximo elemento de un arreglo

```
import numpy as np
def maximoElementoArray(vector):
  maximo = 0
  if len(vector) == 1:
    maximo = max(maximo, vector)
  else:
    maximo = max(maximo, maximoElementoArray(vector[1:]))
  return maximo

maximoElementoArray(np.array([1, 3, 5, 9]))
```

```
#arreglo = [2,1,3,10,5,4,7]
#max(arreglo) = arreglo[0]                                si len(arreglo) = 1
#               maximo(arreglo[0] y max(arreglo[1:]))     si len(arreglo) > 1

#Suponemos que vector no esta vacio
def maximoArreglo(vector:np.array)->int:
  maximo = None
  if len(vector) == 1:
    maximo = vector[0]
  else:
    maximoResto = maximoArreglo(vector[1:]) #maximoArreglo([1,3,10,5,4,7])...maximoArreglo([10,5,4,7])
    if vector[0] > maximoResto:
      maximo = vector[0]
    else:
      maximo = maximoResto
  return maximo

maximoArreglo(np.array([3, 6, 8]))
```
