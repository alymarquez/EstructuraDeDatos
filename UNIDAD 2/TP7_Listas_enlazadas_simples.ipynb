```
#Lista recursiva
import copy as cp

class ListaRec:
  class __NodoLista:
    def __init__(self, dato):
      self.dato = dato
      self.siguiente = None
    def tieneSiguiente(self)->bool:
      return self.siguiente != None
    def __repr__(self)->str:
      salida = ""
      if self != None:
        salida = f" -> {self.dato}"
        if self.tieneSiguiente():
          salida += str(self.siguiente)
      return salida

  def __init__(self):
    self.__primero = None

  def estaVacia(self)->bool:
    return self.__primero == None

  def __repr__(self)->str:
    salida = "primero"
    if not self.estaVacia():
      salida += str(self.__primero)
    return salida + " -|"
```

```
class ListaRec(ListaRec):
  def agregarAlFinal(self, dato:any)->None:
    nodoNuevo = ListaRec.__NodoLista(dato)
    if self.estaVacia():
      self.__primero = nodoNuevo
    else:
      self.__primero.agregarAlFinalNodo(nodoNuevo)

  class __NodoLista(ListaRec.__NodoLista):
    def agregarAlFinalNodo(self, nodoNuevo)->None:
      if not self.tieneSiguiente():
        self.siguiente = nodoNuevo
      else:
        self.siguiente.agregarAlFinalNodo(nodoNuevo)

l1 = ListaRec()
l1.agregarAlFinal(4);l1.agregarAlFinal(1);l1.agregarAlFinal(3);

print(l1)
```

Por ejemplo:

```
class Lista (Lista):

  def eliminarOcurrencias(self,elem):
    cantDel = 0
    if not self.estaVacia():
      #Eliminamos apariciones del elem al inicio
      while not self.estaVacia() and self.__primero.dato == elem:
        self.__primero = self.__primero.siguiente
        cantDel +=1

      #Eliminamos el resto a apartir de la posicion 1
      if not self.estaVacia():
        nodoAux = self.__primero
        while nodoAux.tieneSiguiente():
          if nodoAux.siguiente.dato == elem:
            #Elimino el siguiente y no me corro, porque cambio el siguiente y no quiero
            nodoAux.siguiente = nodoAux.siguiente.siguiente
            cantDel += 1
          else:
            #No elimino entonces solo me muevo al siguiente
            nodoAux = nodoAux.siguiente
    return cantDel
```

**lista1.reemplazar(2,3)** #Reemplaza todos los 2 por 3

```
class Lista(Lista):
  def reemplazar(self, elemAReemplazar, elemNuevo):
    nodoAux = self.__primero
    while nodoAux != None:
      if nodoAux.dato == elemAReemplazar:
        nodoAux.dato = elemNuevo
      nodoAux = nodoAux.siguiente

l1 = Lista()
l1.agregarAlFinal(2);l1.agregarAlFinal(1);l1.agregarAlFinal(4);l1.agregarAlFinal(8);l1.agregarAlFinal(9);l1.agregarAlFinal(2)
print(l1); l1.reemplazar(2, 3); print(l1)
```
